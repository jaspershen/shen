```python
#
# random-life.py
#

from random import random
import matplotlib.pyplot as plt
```

```python
# coordinate: x and y location
# population: red or blue
cNow = ['red', 'blue']
xNow = [-50,  50]
yNow = [ 50, -50]
```

```python
#
# Each generation there can be death or continued life for existing
# cells. There can also be birth.  For pDeath change 1000 to 100 to
# increase the chance of death. 
#
# There are two populations of cell, red and blue. On birth the cell
# is the same color as the cell that continues to live.
#

LIFE = 25               # increase for more generations
CHANCE_OF_DEATH = 1000  # decrease for increased chance a cell dies
```

```python
plt.ylabel('y space')
plt.xlabel('x space')
plt.title('Life')

for generation in range(LIFE):
    numNow = len(cNow)
    pDeath = numNow / CHANCE_OF_DEATH
    xNext = []
    yNext = []
    cNext = []
    for i in range(numNow):
        if random() < pDeath:   # death
            continue
        xNext.append(xNow[i]) # lives
        yNext.append(yNow[i])        
        cNext.append(cNow[i])
        if random() > .5:       # birth
            xNext.append(xNow[i] + 20 * (random() - 0.5))
            yNext.append(yNow[i] + 20 * (random() - 0.5)) 
            cNext.append(cNow[i])
    xNow = xNext
    yNow = yNext
    cNow = cNext
    
    if generation % 5 == 0:
        print('generation', generation)
        print('numNow:', numNow)
        for k in range(len(xNow)):
            plt.plot(xNow[k], yNow[k], 'o', color=cNow[k], alpha=.55)
        plt.show()
```

```python
#
# simple-facs-display-2.py
#

import matplotlib.pyplot as plt
import math

x_values = []
y_values = []
c_values = []

data = {}
    
#ax = plt.gca() 
```

```python
# scatter plot of from FACS data
# CD4 value by CD8 filtered by CD3
CD3 = 4
CD4 = 12
CD8 = 13

# FACS data
with open('NormalMarrowFlow.txt', 'r') as f:
    header_line = next(f).strip().split('\t')
    print(header_line)
    
    for line in f:
        columns = line.strip().split('\t')
        cd3 = float(columns[CD3])
        cd4 = float(columns[CD4])
        cd8 = float(columns[CD8])
        # reduce data space using log base 10, math.log10
        if cd3 > 10:
            x = round(math.log10(max(0.5, cd4))*20)
            y = round(math.log10(max(0.5, cd8))*20)

            # collect the number of cells with a particular CD4, CD8 value
            if (x,y) in data:
                data[(x,y)] += 1
            else:
                data[(x,y)] = 1
```

```python
# dictionary with tuple (x,y) as key to count as value
data
```

```python
# generate color list
colors = []
# size = 50
s = 50
# number of items in data
num_data = len(data)

for (x2,y2) in data:
    x_values.append(10**(x2/20))
    y_values.append(10**(y2/20))
    c_values.append(min(max(data[(x2, y2)]*75, 0), num_data))
```

```python
# plot with axis labels, title, and grid
fig, ax = plt.subplots(figsize=(12,9))
im = ax.scatter(x_values, y_values, s=s, c=c_values, alpha=1.)

fig.colorbar(im, ax=ax)
im.set_clim(0.0, num_data)
plt.xscale('log')
plt.yscale('log')
plt.title('CD4 vs CD8, when CD3 > 10.0')
plt.grid(True)
plt.xlabel('CD4, log count')
plt.ylabel('CD8, log count')
plt.show()
```

```python
# change the background color of image

import matplotlib.pyplot as plt
import matplotlib.image as mpimg

#img1 = mpimg.imread('Spherocytes.jpg')

# img1 is list of values = the width of the image
# each value of img1 is a list = height of the image
w = len(img)
h = len(img[0])

print('width (x-axis)', w, 'height (y-axis)', h)
```

```python
# red, green, blue (RGD)
print(img[x][y])
```

```python
# initialize a blank image that is the same size at img1
# each value of the y array needs an [red, green, blue] value
# RGB values have a range between 0 - 255 
#  [full color = 0, no color = 255]

img2 = []

for x in range(w):
    img2.append([])
    for y in range(h):
        img2[x].append([0, 0, 0])

MAXGREEN = 200

# if the green value of a pixel is > 200 set value to white

for x in range(w):
    for y in range(h):
        r, g, b = img[x][y]
        if g > MAXGREEN:
            img2[x][y] = [255, 255, 255]
        else:
            img2[x][y] = [r, g, b]      
```

```python
fig = plt.figure()

imgplot = plt.imshow(img2)

# save image to file
# plot_output_filename = 'Spherocytes-changed-background.png'
# plt.savefig(plot_output_filename, dpi=200)

plt.show()
```
